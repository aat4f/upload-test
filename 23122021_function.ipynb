```python
#function to filter out integers from a dataset

def test6(a):
    n = []
    if type(a) == list:
        for i in a:
            if type(i) == int:
                n.append(i)
            elif type(i) == list or type(i) == tuple:
                for j in i:
                    if type(j) == int:
                        n.append(j)
    
    return n            #'print n' or 'return n'? here, I want the function to return a list, not a NoneType
```

```python
#function to make a triangle with rows as input

def test10(n):
    #function description (shift+tab)
    """this is a fun  to create a triangle
    you can give input as int
    and you will be able to get output as a triangle"""
    
    for i in range (0,n):
        for j in range (0,i+1):
            print("*",end=" ")
        print("\r")
```
